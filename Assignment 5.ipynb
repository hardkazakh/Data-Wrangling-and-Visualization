# GPA of each student

```python
# I found points of each student
scores = []
for i in range(32):
    scores.append((as1[i]+as2[i]+as3[i]+as4[i])*0.05+quiz[i]*0.3+midterm[i]*0.5)

# I put them their traditional name
for i in range(32):
    if scores[i] >= 90:
        print(df['Student'][i] + ": " + str(scores[i]) + " is Excellent")
    elif 70 <= scores[i] < 90:
        print(df['Student'][i] + ": " + str(scores[i]) + " is Good")
    elif 50 <= scores[i] < 70:
        print(df['Student'][i] + ": " + str(scores[i]) + " is Satisfactory")
    elif 50 > scores[i]:
        print(df['Student'][i] + ": " + str(scores[i]) + ' is Fall')
```

```python
# I found their LGs and GPA points
for i in range(32):
    if scores[i] >= 95:
        print(df['Student'][i] + " " + str(scores[i]) + ": " + " LG is A, Point - 4")
    elif 90 <= scores[i] < 95:
        print(df['Student'][i] + " " + str(scores[i]) + ": " + " LG is A-, Point - 3.67")
    elif 85 <= scores[i] < 90:
        print(df['Student'][i] + " " + str(scores[i]) + ": " + " LG - B+, Point - 3.33")
    elif 80 <= scores[i] < 85:
        print(df['Student'][i] + " " + str(scores[i]) + ": " + " LG - B, Point - 3")
    elif 75 <= scores[i] < 80:
        print(df['Student'][i] + " " + str(scores[i]) + ": " + " LG - B-, Point - 2.67")
    elif 70 <= scores[i] < 75:
        print(df['Student'][i] + " " + str(scores[i]) + ": " + " LG - C+, Point - 2.33")
    elif 65 <= scores[i] < 70:
        print(df['Student'][i] + " " + str(scores[i]) + ": " + " LG - C, Point - 2")
    elif 60 <= scores[i] < 65:
        print(df['Student'][i] + " " + str(scores[i]) + ": " + " LG - C-, Point - 1.67")
    elif 55 <= scores[i] < 60:
        print(df['Student'][i] + " " + str(scores[i]) + ": " + " LG - D+, Point - 1.33")
    elif 50 <= scores[i] < 55:
        print(df['Student'][i] + " " + str(scores[i]) + ": " + " LG - D, Point - 1")
    else:
        print(df['Student'][i] + " " + str(scores[i]) + ": " + " LG - F, Point - 0")
```

# Grouping with GPA

```python
# Firstly, I created empty lists
excellent = []
good = []
satisfactory = []
fall = []
```

```python
# I grouped them by points
for i in range(32):
    if scores[i] >= 90:
        excellent.append(df['Student'][i] + ": " + str(scores[i]))
    elif 70 <= scores[i] < 90:
        good.append(df['Student'][i] + ": " + str(scores[i]))
    elif 50 <= scores[i] < 70:
        satisfactory.append(df['Student'][i] + ": " + str(scores[i]))
    elif 50 > scores[i]:
        fall.append(df['Student'][i] + ": " + str(scores[i]))
```

```python
# print all groups
print("<<EXCELLENT>>")
for i in excellent:
    print(i)
print()

print("<<GOOD>>")
for i in good:
    print(i)
print()

print("<<SATISFACTORY>>")
for i in satisfactory:
    print(i)
print()

print("<<FALL>>")
for i in fall:
    print(i)
print()
```
