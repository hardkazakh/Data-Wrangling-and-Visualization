__5. Copy existing dataframe and delete all rows that repeats more than twice by Athlete column__

```python
df3 = df.copy()
athletes = []
indexes_more_than_two = []
counter = 0
for i in df3['Athlete']:
    if athletes.count(i) > 1:
        indexes_more_than_two.append(counter)
    else:
        athletes.append(i)
    counter += 1

df3.drop(index=indexes_more_than_two)
```

__6. The dataset has elements of categorical data in the “Sport” column. Create a new dataframe with convertion of Sport column by one hot encoded column and left join this dataset to our main table.__

```python
sports = []
for i in df['Sport']:
    if i in sports:
        continue
    else:
        sports.append(i)    
sports.sort()

df['Numeric'] = 0
df.Numeric = pd.Categorical(df.Sport, sports, ordered=True)
df.Numeric = df.Numeric.cat.codes
df
```
